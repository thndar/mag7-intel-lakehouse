#### 1. Get environment variables and initialize BigQuery client

```python
import os
from dotenv import load_dotenv
load_dotenv()  # loads .env in current working dir
from google.cloud import bigquery

PROJECT_ID = os.getenv("GCP_PROJECT_ID")
client = bigquery.Client(project=PROJECT_ID)

print("GOOGLE_APPLICATION_CREDENTIALS =", os.getenv("GOOGLE_APPLICATION_CREDENTIALS"))
print("GCP_PROJECT_ID =", PROJECT_ID)
```

```python
from pathlib import Path

PROJECT_ROOT = Path("/home/shpang/devs/ntu/projects/mag7_intel")
SQL_PATH = PROJECT_ROOT / "research" / "sql" / "research_join.sql"

QUERY = SQL_PATH.read_text().replace("{{project}}", PROJECT_ID)

df = client.query(QUERY).to_dataframe()
df.head(), df.shape
```
